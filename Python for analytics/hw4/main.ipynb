```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

kc_house_data = pd.read_csv('kc-house-data.csv')
```

```python
plt.figure(figsize=(10, 6))
# plt.hist(kc_house_data['price'], bins=50, color='skyblue', edgecolor='black')
plt.hist(kc_house_data['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость (USD)')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
```

```python
#  2.1 Построение графика для распределения домов от наличия вида на набережную
waterfront_count = kc_house_data['waterfront'].value_counts()


names = ['Без вида на набережную', 'С видом на набережную']
values = waterfront_count.values
plt.pie(values, labels=names, autopct='%.1f%%', startangle=100);
```

```python
# Построение scatter plot для визуализации домов с учетом наличия вида на набережную
plt.figure(figsize=(10, 6))

# Дома без вида на набережную
plt.scatter(kc_house_data[kc_house_data['waterfront'] == 0]['long'], 
            kc_house_data[kc_house_data['waterfront'] == 0]['lat'], 
            color='blue', alpha=0.5, label='Без вида на набережную')

# Дома с видом на набережную
plt.scatter(kc_house_data[kc_house_data['waterfront'] == 1]['long'], 
            kc_house_data[kc_house_data['waterfront'] == 1]['lat'], 
            color='red', alpha=0.5, label='С видом на набережную')

plt.title('Распределение домов по наличию вида на набережную')
plt.xlabel('Долгота')
plt.ylabel('Широта')
plt.legend()
plt.show()
```

Дома с видом на набережную обычно встречаются реже, но значительно выше ценятся на рынке из-за своего уникального расположения и видов, которые они предлагают.

```python
# 2.2 Распределение домов по количеству этажей

floors_count = kc_house_data['floors'].value_counts().sort_index()

names = floors_count.index.astype(str)
values = floors_count.values

plt.figure(figsize=(10, 6))
plt.bar(names, values, color='orange')
plt.title('Распределение количества этажей в домах')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.grid(axis='y', linestyle='--')
plt.show()
```

Большинство домов имеют 1 или 2 этажа. Дома с одним этажом часто предпочитаются старшим поколениям или семьям с маленькими детьми из-за легкости доступа. Дома с двумя и более этажами могут предложить больше жилого пространства на меньшем участке земли.

```python
# 2.3 Распределение домов по состоянию

condition_count = kc_house_data['condition'].value_counts().sort_index()

names = condition_count.index.astype(str)
values = condition_count.values

plt.figure(figsize=(10, 6))
plt.bar(names, values, color='mediumpurple')
plt.title('Распределение домов по состоянию')
plt.xlabel('Состояние')
plt.ylabel('Количество домов')
plt.grid(axis='y', linestyle='--')
plt.show()
```

Состояние дома играет важную роль в его стоимости. Дома в хорошем или отличном состоянии, как правило, стоят дороже, чем дома, требующие значительного ремонта или обновления.

## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

#### Жилая площадь (sqft_living):
Линейная диаграмма или диаграмма рассеяния (scatter plot) между жилой площадью и стоимостью может помочь выявить взаимосвязь между размером дома и его стоимостью

```python
plt.figure(figsize=(10, 6))
plt.scatter(kc_house_data['sqft_living'], kc_house_data['price'], alpha=0.5)
plt.title('Влияние жилой площади на стоимость дома')
plt.xlabel('Жилая площадь (кв. футы)')
plt.ylabel('Стоимость дома (USD)')
plt.grid(True)
plt.show()
```

#### Количество спален (bedrooms): 
Столбчатая диаграмма средней стоимости домов в зависимости от количества спален может показать, как количество спален влияет на цену недвижимости.

```python
plt.figure(figsize=(10, 6))
kc_house_data.groupby('bedrooms')['price'].mean().plot(kind='bar', color='coral')
plt.title('Средняя стоимость дома в зависимости от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Средняя стоимость дома (USD)')
plt.grid(axis='y')
plt.show()
```

#### Количество ванных комнат (bathrooms): 

Аналогично количеству спален, столбчатая диаграмма средней стоимости может иллюстрировать влияние количества ванных комнат на стоимость

```python
import numpy as np

bathroom_price_mean = kc_house_data.groupby('bathrooms')['price'].mean().reset_index()

plt.figure(figsize=(12, 8))
plt.bar(bathroom_price_mean['bathrooms'], bathroom_price_mean['price'], color='cadetblue')
plt.title('Средняя стоимость домов в зависимости от количества ванных комнат')
plt.xlabel('Количество ванных комнат')
plt.ylabel('Средняя стоимость дома (USD)')
plt.xticks(np.arange(min(bathroom_price_mean['bathrooms']), max(bathroom_price_mean['bathrooms'])+1, 0.5), rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

#### Вид на набережную (waterfront): 
Сравнение средней стоимости домов с видом на набережную и без может быть представлено через столбчатую диаграмму, подчеркивая влияние этой характеристики на ценность недвижимости

```python
plt.figure(figsize=(10, 6))
kc_house_data.groupby('waterfront')['price'].mean().plot(kind='bar', color=['skyblue', 'navy'])
plt.title('Средняя стоимость дома с видом и без вида на набережную')
plt.xlabel('Вид на набережную')
plt.ylabel('Средняя стоимость дома (USD)')
plt.xticks(ticks=[0, 1], labels=['Без вида', 'С видом'], rotation=0)
plt.show()
```

#### Год постройки (yr_built): 
Линейная диаграмма или диаграмма рассеяния, показывающая стоимость домов в зависимости от года постройки, может выявить, предпочитают ли покупатели новые или старые дома и как это влияет на цену.

```python
plt.figure(figsize=(12, 6))
kc_house_data.groupby('yr_built')['price'].mean().plot(color='purple')
plt.title('Средняя стоимость дома в зависимости от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Средняя стоимость дома (USD)')
plt.grid(True)
plt.show()
```
